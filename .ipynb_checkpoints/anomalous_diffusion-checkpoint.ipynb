```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def anomalous_diffusion_2d(num_steps, alpha=1.0, D=1.0, dt=1.0):
    # Generate random displacements for x and y directions
    # The scale factor is adjusted by the diffusion coefficient D and the exponent alpha
    scale = np.sqrt(2 * D * dt**alpha)
    #print(scale, 'a=', alpha, 'dt=', dt, (dt**alpha))
    dx = np.random.normal(loc=0, scale=scale, size=num_steps)
    dy = np.random.normal(loc=0, scale=scale, size=num_steps)
    x = np.cumsum(dx)
    y = np.cumsum(dy)
    return x, y

def compute_msd(x, y):
    """ Compute the Mean Squared Displacement (MSD) of a 2D trajectory. """
    num_steps = len(x)
    msd = np.zeros(num_steps)
    for tau in range(num_steps):
        displacements = (x[tau:] - x[:num_steps-tau])**2 + (y[tau:] - y[:num_steps-tau])**2
        msd[tau] = np.mean(displacements)
    return msd
    
def normal_diffusion_function(t, D):
    """ Function to fit the MSD data: MSD(t) = 4 * D * t (2D case) """
    return 4 * D * t

def fit_normal_coef(msd, dt):
    """ Fit the MSD curve to recover D. """
    time = np.arange(len(msd)) * dt 
    popt, pcov = curve_fit(normal_diffusion_function, time, msd, p0=(1.0))
    D_fit = popt
    return D_fit[0]
    
def anomalous_diffusion_function(t, D, alpha):
    """ Function to fit the MSD data: MSD(t) = 4 * D * t^alpha (2D case) """
    return 4 * D * t**alpha

def fit_anomalous_coef(msd, dt):
    """ Fit the MSD curve to recover D and alpha. """
    time = np.arange(len(msd)) * dt 
    popt, pcov = curve_fit(anomalous_diffusion_function, time, msd, p0=(1.0, 1.0))
    D_fit, alpha_fit = popt
    return D_fit, alpha_fit

# Parameters
num_particles = 1
N = 10  # Number of steps in the random walk
num_frames = 20
alpha = 1       # Anomalous diffusion exponent (1.0 corresponds to normal Brownian motion)
D = 10           # Diffusion coefficient
dt = 0.001          # Time step

num_steps = N*num_frames
time = np.arange(num_steps) * dt
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].set_title(f'{num_particles} particles with {num_steps} steps (D={D} a={alpha})')
axs[0].grid(True)
axs[0].axis('equal')
axs[1].set_title('MSD and Fit')
axs[1].grid(True)

msd = np.zeros((num_particles, num_steps))
for k in range(num_particles):
    x, y = anomalous_diffusion_2d(num_steps, alpha, D, dt)
    print(x[0:4])
    msd[k] = compute_msd(x, y)
    axs[0].plot(x, y, lw=0.25)
    axs[1].plot(time, msd[k], lw=0.25)

mean_curve = np.mean(msd, axis=0)
std_curve = np.std(msd, axis=0)
D_fit_mean = fit_normal_coef(mean_curve, dt)
A_fit_mean, alpha_fit_mean = fit_anomalous_coef(mean_curve, dt)
print('Normal    Diffusion D_fit', D_fit_mean )
print('Anomalous Diffusion D_fit', A_fit_mean, ' alpha ', alpha_fit_mean)

axs[1].plot(time, normal_diffusion_function(time, D_fit_mean), 'r-', label=f'Fit: D={D_fit_mean:.3f}')
axs[1].plot(time, anomalous_diffusion_function(time, A_fit_mean, alpha_fit_mean), 'b-', label=f'Fit: D={A_fit_mean:.3f}, alpha={alpha_fit_mean:.3f}')
axs[1].plot(time, mean_curve, 'k-', label=f'Mean')
axs[1].plot(time, mean_curve + std_curve, 'k--', label='+1 Std Dev')
axs[1].plot(time, mean_curve - std_curve, 'k--', label='-1 Std Dev')

axs[1].legend()
plt.show()
```
