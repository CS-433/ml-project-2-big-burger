```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def anomalous_diffusion_2d(num_steps, alpha=1.0, D=1.0, dt=1.0):
    # Generate random displacements for x and y directions
    # The scale factor is adjusted by the diffusion coefficient D and the exponent alpha
    scale = np.sqrt(2 * D * dt**alpha)
    dx = np.random.normal(loc=0, scale=scale, size=num_steps)
    dy = np.random.normal(loc=0, scale=scale, size=num_steps)
    
    # Calculate the position by cumulative sum
    x = np.cumsum(dx)
    y = np.cumsum(dy)
    
    return x, y

def compute_msd(x, y):
    """ Compute the Mean Squared Displacement (MSD) of a 2D trajectory. """
    num_steps = len(x)
    msd = np.zeros(num_steps)
    
    for tau in range(num_steps):
        displacements = (x[tau:] - x[:num_steps-tau])**2 + (y[tau:] - y[:num_steps-tau])**2
        msd[tau] = np.mean(displacements)
    
    return msd
def normal_diffusion_function(t, D):
    """ Function to fit the MSD data: MSD(t) = 4 * D * t^alpha (2D case) """
    return 4 * D * t

def fit_normal_coef(msd, dt):
    """ Fit the MSD curve to recover D and alpha. """
    time = np.arange(len(msd)) * dt 
    popt, pcov = curve_fit(normal_diffusion_function, time, msd, p0=(1.0))
    D_fit = popt
    return D_fit
    
def anomalous_diffusion_function(t, D, alpha):
    """ Function to fit the MSD data: MSD(t) = 4 * D * t^alpha (2D case) """
    return 4 * D * t**alpha

def fit_anomalous_coef(msd, dt):
    """ Fit the MSD curve to recover D and alpha. """
    time = np.arange(len(msd)) * dt 
    popt, pcov = curve_fit(anomalous_diffusion_function, time, msd, p0=(1.0, 1.0))
    D_fit, alpha_fit = popt
    return D_fit, alpha_fit

# Parameters
num_steps = 10000  # Number of steps in the random walk
alpha = 1.0       # Anomalous diffusion exponent (1.0 corresponds to normal Brownian motion)
D = 1.0           # Diffusion coefficient
dt = 1.0          # Time step

# Generate the 2D anomalous diffusion path
x, y = anomalous_diffusion_2d(num_steps, alpha, D, dt)

# Compute the MSD from the trajectory
msd = compute_msd(x, y)

D_fit = fit_normal_coef(msd, dt)
A_fit, alpha_fit = fit_anomalous_coef(msd, dt)


plt.figure(figsize=(8, 8))
plt.plot(x, y, lw=1)
plt.title(f'2D Anomalous Diffusion (alpha={alpha})')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axis('equal')
plt.show()

# Plot the MSD and the fit
time = np.arange(len(msd)) * dt
plt.figure(figsize=(8, 6))
plt.plot(time, msd, 'o', label='MSD (computed)', markersize=5)
plt.plot(time, normal_diffusion_function(time, D_fit), '-', label=f'Fit: D={D_fit:.3f}')
plt.plot(time, anomalous_diffusion_function(time, A_fit, alpha_fit), '-', label=f'Fit: D={A_fit:.3f}, alpha={alpha_fit:.3f}')
plt.title('Mean Squared Displacement (MSD) and Fit')
plt.xlabel('Time')
plt.ylabel('MSD')
plt.legend()
plt.grid(True)
plt.show()

# Print the fitted parameters
print(f"Recovered parameters: D = {D_fit:.3f}, alpha = {alpha_fit:.3f}")
```
